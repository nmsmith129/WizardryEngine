```python
## Class definition for a map space

class Space:
    def __init__(family, type, location, direction):
        self.__family = family
        self.__type = type
        self.__location = location
        self.__direction = direction

    def getFamily():
        return self.__family

    def getType():
        return self.__type

    def getLocation():
        return self.__location

    def getDirection():
        return self.__direction
```

```python
## Class for Room

class Room:
    def __init__(contents):
        self.__contents = contents
        return self

    def getContents():
        return self.__contents

    def removeContents(removable):
        self.__contents.remove(removable)
        return
```
